#### Add a 5% bonus percentage to the student whose percentage is more than 65%. Update the Percentage column with new values

```python
print(Latest.dtypes)
```

```python
#Since percent is in float datatype and we know for comparison purpose it is needful to get that converted 
Latest['Percent']=Latest['Percent'].astype(int)
```

```python
print(Latest.dtypes)
```

```python

Latest['Percent']=Latest.apply(lambda x:(x['Percent']+(x['Percent']*0.05)) if x['Percent']>65 else x['Percent'],axis=1)
```

```python
#updated percentage
Latest
```

```python

plt.figure(figsize = (12,6))
plt.scatter(x='Max',y='Student_id',data=Latest, c = "r", marker = "*", s = 100, alpha=0.5, linewidths=10,edgecolors="g" )
plt.xlabel('Max',color='r')
plt.ylabel('Student_id',color='r')
plt.show()
```

###  Update Student Id, replace the first letter by "X" in place of "A" where the percentage is less than 60%. Other remaining letters of student id must remain the same. 

```python
Latest['Student_id']=Latest.apply(lambda x:x['Student_id'].replace('A','X') if x['Percent']<60 else x['Student_id'],axis=1)
```
